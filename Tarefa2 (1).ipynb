```python
from dataclasses import dataclass

@dataclass
class Serie:
    titulo: str
    ano_estreia: str
    total_episodios: float
    nota: float
    pagina: str
```

```python
import re
def cria_serie(imdb_li_tag):
    titulo = imdb_li_tag.find_element(By.CLASS_NAME, "ipc-title__text").text #pegamos o texto do elemento
    ano_texto = imdb_li_tag.find_element(By.XPATH, r'./div/div/div/div/div[2]/div[2]/span[1]').text
    # Express√£o regular que captura o primeiro ano (sempre 4 d√≠gitos)
    match = re.search(r'(\d{4})', ano_texto)
    if match:
        ano_estreia = match.group(1)
    total_episodios = imdb_li_tag.find_element(By.XPATH, r'./div/div/div/div/div[2]/div[2]/span[2]').text
    nota = imdb_li_tag.find_element(By.CLASS_NAME, "ipc-rating-star--rating").text 
    pagina = imdb_li_tag.find_element(By.CLASS_NAME, "ipc-title-link-wrapper").get_attribute("href")

    return Serie(titulo,ano_estreia,total_episodios,nota,pagina)
```

```python
lista_de_series = []
for serie_tag in lista_series:
    serie = cria_serie(serie_tag)
    lista_de_series.append(serie)
```
